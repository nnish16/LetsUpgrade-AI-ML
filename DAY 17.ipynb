```python
import pandas as pd
import seaborn as sns
import numpy as np
import os
```

```python
data = pd.read_csv("mtcars.csv")
data.head(25)
```

```python
data.describe()
```

```python
data.info()
data.shape
```

```python
sns.boxplot(x = data["hp"], data = data)
```

```python
Mean = np.mean(data["hp"])

SD = np.std(data["hp"])
```

```python
print("The mean of the data is :-",Mean)
print("The standard deviation is :-",SD)
```

```python
def check_outlier(data):
    threshold = 2
    outlier = []
    for i in data:
        z_score = (i - Mean)/SD
        if z_score > threshold:
            outlier.append(i)
    print(outlier)
```

```python
check_outlier(data["hp"])
```

Q2 Using the Boston data set from sklearn, find out the Outlier and treat them

```python
from sklearn.datasets import load_boston

boston = load_boston()
x = boston.data
y = boston.target
columns = boston.feature_names

boston_data = pd.DataFrame(x,columns=columns)
boston_data.head(25)
```

```python
boston_data.info()
```

```python
boston_data.describe()
boston_data.shape
```

```python
sns.boxplot(boston_data["DIS"])
```

```python
Q1 = boston_data["DIS"].quantile(0.25)
Q3 = boston_data["DIS"].quantile(0.75)

IQR = Q3 - Q1

print(Q1)
print(Q3)
print(IQR)
```

```python
Upper_whisker = Q3 + (1.5*IQR)
Lower_whisker = Q1 - (1.5*IQR)


print(Upper_whisker)
print(Lower_whisker)
```

```python
for i in boston_data["DIS"]:
    if i > Upper_whisker:
        print(i)
    elif i < Lower_whisker:
        print(i)
```

```python
new_data = boston_data[boston_data["DIS"] < Upper_whisker]

new_data.shape
```
